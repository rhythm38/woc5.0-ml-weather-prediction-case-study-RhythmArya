```python
df_with_missing_values_columns = df[['Creator Name', 'Creator Gender','Language of the Video','Date of the Last Comment','Subtitle','No of Likes']]
df_with_missing_values_columns
```

```python
#example of dropping NULL values
df1 = df_with_missing_values_columns.dropna(subset=['Language of the Video'])
df1
```

As we can see in above dataframe after dropping one row having NaN value,no of row reduced by one.

```python
df_with_missing_values_columns["Creator Gender"].fillna("Female", inplace = True)
df_with_missing_values_columns
```

```python
df_with_missing_values_columns.isna().sum()
```

```python
df_with_missing_values_columns
```

The ffill() method replaces the NULL values with the value from the previous row (or previous column, if the axis parameter is set to 'columns').Similarly bfill, backfill and pad methods can also be used. 

```python
df_with_missing_values_columns["Language of the Video"].fillna(method ='ffill', inplace = True)
df_with_missing_values_columns
```

```python
df3 = df_with_missing_values_columns.fillna(method='ffill')
df3
```

```python
df3.isna().sum()
```
